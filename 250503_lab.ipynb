```python
# 老師的解法:利用函式抓取多個頁面的爬蟲資料
import requests
import pandas as pd
import bs4 as bs

url = "https://24h.pchome.com.tw/search/?q=%E5%90%B9%E9%A2%A8%E6%A9%9F"
# req.urlretrieve(url, "xx.html")
resp = requests.get(url)
html = bs.BeautifulSoup(resp.text)
# print(html)

# DataFrame: [{}, {}, {}]
table = []
prod_list = html.find_all("a", {"class":"c-prodInfoV2__link"})
for prod in prod_list:
    prod_url = "https://24h.pchome.com.tw" + prod["href"]
    print(prod_url)
    data = analaze_page(prod_url)
    table.append(data)

df = pd.json_normalize(table)
df.to_csv("pchome.csv", encoding="utf-8")
```

```python
# 老師的解法:定義成函式
import requests
import bs4 as bs
import os

def analaze_page(url):
    # url = "https://24h.pchome.com.tw/prod/DMBP01-A900GPN0L"
    resp = requests.get(url)
    html = bs.BeautifulSoup(resp.text)
    # 名稱/品牌
    main_name = html.find("div", {"class":"o-prodMainName--prodNick"})
    brand = main_name.find("a")
    if not brand == None:
        brand_text = brand.text
    else:
        brand_text = "-"
    goods_name = main_name.find("h1")
    goods_name_text = goods_name.text
    # 價格
    price = html.find("div", {"class":"o-prodPrice__price"})
    price_text = price.text
    # 圖片
    goods_img = html.find("div", {"class":"c-radiusPhotoImage__img"}).find("img")
    goods_img_src = goods_img["src"]
    if not "https" in goods_img_src:
        goods_img_src = "https:" + goods_img_src

    # print(brand_text)
    # print(goods_name_text)
    # print(price_text)
    # print(goods_img_src)

    # download img
    # 如果可以, 當然用urlretrieve
    dn = "pchome"
    if not os.path.exists(dn):
        os.makedirs(dn)
    fn = url.split("/")[-1]
    fn = dn + "/" + fn + ".jpg"
    img_resp = requests.get(goods_img_src, stream=True)
    with open(fn, "wb") as f:
        f.write(img_resp.raw.read())

    data = {
        "brand": brand_text,
        "name": goods_name_text,
        "price": price_text,
        "img": goods_img_src,
        "img_path":fn
    }
    return data
```
